```python
import json
import numpy as np
import matplotlib.pyplot as plt
 
```

```python
# Read all the logs inside the folder logs
import os
import sys
import re

FOLDER = "logs"

def read_json_logs(folder):
    logs = []
    for filename in os.listdir(folder):
        if filename.endswith(".json"):
            with open(os.path.join(folder, filename), 'r') as f:
                logs.append(json.load(f))
    return logs

logs = read_json_logs(FOLDER)
```

```python
def plot_logs(logs, keys):
    for key in keys:
        to_plot = sorted([(i["epoch"],np.mean(i[key])) for i in logs], key=lambda x: x[0])
        plt.plot([i[0] for i in to_plot], [i[1] for i in to_plot], label=key)
        plt.legend()
    plt.show()
    
keys = [["batch_train_losses", "batch_val_losses"], ["batch_train_accuracies", "batch_val_accuracies"]]
for key in keys:
    plot_logs(logs, key)
```
